```python
import fitz
import pandas as pd
```

```python
document_path = r'C:\Users\sudip\Workspace\euclid_assignment\data\harry-potter-book-collection-1-4.pdf'
```

```python
doc = fitz.open(document_path)
```

```python
def check_chapter_in_page(text):
    if 'CHAPTER' in text or 'C H A P T E R' in text:
        return True
    else:
        return False

def split_chapter(text):
    if 'CHAPTER' in text:
        post_chapter_text = 'CHAPTER' + text.split('CHAPTER')[1]
        pre_chapter_text = text.split('CHAPTER')[0]
    else:
        post_chapter_text = 'CHAPTER' + text.split('C H A P T E R')[1]
        pre_chapter_text = text.split('C H A P T E R')[0]
    
    return pre_chapter_text, post_chapter_text
    
```

```python
chapter_count = 0
chapter_text = ''

book_name_list = []
chapter_list = []
chapter_text_list = []

for idx, page in enumerate(doc):
    text = page.get_text()

    if text == '':
        print('Start of a new book.')

        if chapter_text != '':
            chapter_list.append(chapter_count)
            chapter_text_list.append(chapter_text)
            book_name_list.extend([book_name]*chapter_count)

        chapter_count = 0
        chapter_text = ''
        book_name = doc[idx+1].get_text()
        # book_name_list.append(book_name)

        continue

    if not check_chapter_in_page(text) and chapter_count == 0: # FOR BEFORE THE FIRST CHAPTER, preface and table of contents
        continue

    elif check_chapter_in_page(text) and chapter_count == 0: # page contains first chapter
        chapter_count += 1
        chapter_text += split_chapter(text)[1]
    
    elif check_chapter_in_page(text) and chapter_count != 0: # page contains starting of other chapters
        chapter_text += split_chapter(text)[0]

        chapter_list.append(chapter_count)
        chapter_text_list.append(chapter_text)

        chapter_count += 1
        chapter_text = split_chapter(text)[1]
    
    else:
        chapter_text += text 

chapter_list.append(chapter_count)
chapter_text_list.append(chapter_text.split('\nNOT FOR SALE')[0])
book_name_list.extend([book_name]*chapter_count)
```

```python
book_name_list = [x.replace('\n', '').replace('J.K. Rowling', '') for x in book_name_list]
```

```python
data = {'book':book_name_list,
'chapter': chapter_list,
'text': chapter_text_list}
```

```python
df = pd.DataFrame(data)
```
